```python
import pandas as pd
import numpy as np

# Visualization libraries
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotnine as p9
```

```python
# 1st. Method: Object Oriented Style (OO-style)

x = np.linspace(0, 5, 100 )  # Sample data, 100 points between 0 and 5.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(10, 6))

ax.grid(True) # Add a grid

ax.plot(x, np.sin(x), label='Sine function ')  # Plot some data on the axes. Can use LaTeX notation.
ax.plot(x, x, label='linear ')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic ')  # ... and some more.
ax.plot(x, x-x**3/6.0, linestyle='dashed', label='')
ax.set_xlim([0, np.pi/2]) # We select limits for plotting
ax.set_ylim([0, 1.5])

# Optional: We can define where and what the xtics can be
ax.set_xticks([0, 0.125*np.pi, 0.25*np.pi, 0.375*np.pi, np.pi/2]) # values
ax.set_xticklabels(['', '','','','']) # labels
ax.set_xlabel('')  # Add an x-label to the axes.
ax.set_ylabel('')  # Add a y-label to the axes.
ax.set_title("Simple function plots")  # Add a title to the axes.
ax.legend();  # Add a legend.
```
